```python
master_df = pd.read_csv("/Users/atharv/Projects/My Projects/ movie_recommendation/data/processed/master_dataset.csv")
master_df.head()
```

```python
master_df["text_features"] = (master_df["genres"].fillna("")+ " " + master_df["tag"].fillna("") + " " + master_df["genome_tags"].fillna(""))
```

```python
master_df["text_features"][0]
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " " , text)
    return text.strip()

master_df["text_features"]= master_df["text_features"].apply(clean_text)
```
